```python
#split into feature and label
feature = df[["Math","Reading"]]
label = df["Writing"]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets (80% training, 20% testing)
X_train, X_test, Y_train, Y_test = train_test_split(feature, label, test_size=0.2, random_state=42)
```

```python
import numpy as np

# Define the cost function
def cost_function(X, Y, W):
    """
    This function finds the Mean Square Error.
    Arguments:
    X -- Feature matrix
    Y -- Target vector
    W -- Weight vector

    Returns:
    cost -- The mean square error
    """
    m = len(Y)  # Number of training examples
    Y_pred = np.dot(X, W)  # Predicted values
    cost = (1 / (2 * m)) * np.sum((Y_pred - Y) ** 2)  # MSE cost function
    return cost
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    """
    Perform gradient descent to optimize the parameters of a linear regression model.
    Arguments:
    X -- Feature matrix
    Y -- Target vector
    W -- Initial weight vector
    alpha -- Learning rate
    iterations -- Number of iterations

    Returns:
    W_update -- Final optimized weight vector
    cost_history -- History of cost values over iterations
    """
    m = len(Y)  # Number of training examples
    cost_history = []  # To store cost history

    for iteration in range(iterations):
        Y_pred = np.dot(X, W)  # Compute predictions
        loss = Y_pred - Y  # Difference between predicted and actual
        gradient = (1 / m) * np.dot(X.T, loss)  # Compute gradient
        W -= alpha * gradient  # Update weights

        # Compute and store cost
        cost = cost_function(X, Y, W)
        cost_history.append(cost)

    return W, cost_history
```

```python
# Generate random test data
np.random.seed(0) # For reproducibility
X = np.random.rand(100, 3) # 100 samples, 3 features
Y = np.random.rand(100)
W = np.random.rand(3) # Initial guess for parameters
# Set hyperparameters
alpha = 0.01
iterations = 1000
# Test the gradient_descent function
final_params, cost_history = gradient_descent(X, Y, W, alpha, iterations)
# Print the final parameters and cost history
print("Final Parameters:", final_params)
print("Cost History:", cost_history)
```

```python
def rmse(Y, Y_pred):
    """
    This function calculates the Root Mean Squared Error.
    Arguments:
    Y -- Actual values
    Y_pred -- Predicted values

    Returns:
    rmse -- Root Mean Squared Error
    """
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

```python
def r2(Y, Y_pred):
    """
    This function calculates the R-Squared error.
    Arguments:
    Y -- Actual values
    Y_pred -- Predicted values

    Returns:
    r2 -- R-Squared value
    """
    ss_total = np.sum((Y - np.mean(Y)) ** 2)  # Total sum of squares
    ss_residual = np.sum((Y - Y_pred) ** 2)  # Residual sum of squares
    return 1 - (ss_residual / ss_total)  # R-squared
```

```python
# Step 4: Initialize weights (W) to zeros, learning rate and number of iterations
W = np.zeros(X_train.shape[1])  # Initialize weights (for two features)
alpha = 0.01  # Learning rate
iterations = 1000  # Number of iterations for gradient descent

# Step 5: Perform Gradient Descent
W_optimal, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

# Step 6: Make predictions on the test set
Y_pred = np.dot(X_test, W_optimal)

# Step 7: Evaluate the model using RMSE and R-Squared
model_rmse = rmse(Y_test, Y_pred)
model_r2 = r2(Y_test, Y_pred)

# Step 8: Output the results
print("Final Weights:", W_optimal)
print("Cost History (First 10 iterations):", cost_history[:10])
print("RMSE on Test Set:", model_rmse)
print("R-Squared on Test Set:", model_r2)
```
