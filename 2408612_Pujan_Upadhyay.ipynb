```python
#1. Column named Happiness_Category that categorize based on their score
# low - score < 4
#medium - 4 <= score <=6
#high - score >6


#function to categorize the happiness
def categorize_happiness(score):
  '''
    this function categorize_happiness will take parameter score from the dataset and
    compare it as:
    if score if less than 4 then in new column of dataset in that particular row 'Low' will be placed
    else if score if greater than or equals to 4 and less than orr equals to 6then in new column of dataset in that particular row 'Medium' will be placed
    else in new column of dataset in that particular row 'High' will be placed
  '''
  if score < 4:
    return 'Low'
  elif 4<= score <=6:
    return 'Medium'
  else:
    return 'High'

World_Happiness_Report['Happiness_Category'] = World_Happiness_Report['score'].apply(categorize_happiness)
World_Happiness_Report
```

# Task- 3: Outlier Dection:

```python
#1. Outliers countires in South Asia based on their Score and GDP per capita
#2. Define outliers using IQR rule

#function for outliers detection
def outliers_find(data,column):
  '''
    the function is used to calculate the outilers in data and return lower bound and upper bound of the data.
    first we calculate first and third quartile using the library function .quantile.
    Then calculate IQR(Inter Quartile Range) as Q3-Q1
    At last calculate the lower bound = Q1 - 1.5*IQR
       and  upper bound = Q3 + 1.5*IQR
  '''
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)

  IQR = Q3-Q1

  upper_bound = Q3 + 1.5*IQR
  lower_bound = Q1 - 1.5*IQR

  return upper_bound, lower_bound
```

```python
#calling outliers_find function to calculate upper bound and lower bound

#for score column
upper_score, lower_score = outliers_find(Dataset_South_Asian,'score')
#for GDP column
upper_gdp, lower_gdp = outliers_find(Dataset_South_Asian,'Log GDP per capita')
```

```python
#identifying outliers data

#for score
outilers_score = Dataset_South_Asian[(Dataset_South_Asian['score']<lower_score) | (Dataset_South_Asian['score']>upper_score) ]

#for GDP
outilers_GDP = Dataset_South_Asian[(Dataset_South_Asian['Log GDP per capita']<lower_gdp) | (Dataset_South_Asian['Log GDP per capita']>upper_gdp) ]

#combine outliers
outliers = pd.concat([outilers_score,outilers_GDP])
```
