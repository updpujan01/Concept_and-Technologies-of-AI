# **Exercise 3.3**

```python
#total sum of fare
t_sum = df['Fare'].sum()
```

```python
#total fare of p1 class
c1 = df[df['Pclass']==1]['Fare'].sum()
```

```python
c2 = df[df['Pclass']==2]['Fare'].sum()
```

```python
c3 = df[df['Pclass']==3]['Fare'].sum()
```

```python
all_fare_class = [c1,c2,c3]
```

```python
fare_proportion_by_class = [fare / t_sum * 100 for fare in all_fare_class]
fare_proportion_by_class
```

# **3.3 second part**

```python
df['age_group'] = pd.cut(df['Age'],bins=[0,18,65,float('inf')],labels=["child","adult","senior"])
df
```

```python
total_passengers = len(df)
total_passengers
```

```python
age_group_counts = df['age_group'].value_counts()
age_group_counts
```

```python
age_group_proportion = (age_group_counts / total_passengers) * 100
age_group_proportion
```
